# Step: 5.Handle the Missing values.

```python
# Checking the null values in dataset
data.isna()
```

```python
data.isnull().any()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
# if we have null values in numerical column we replace it with Mean/Median
#if we have null values in categorical column we replace it with Mode
```

```python
data['age'].fillna(data['age'].mean(),inplace=True)
```

```python
data['embarked'].fillna(data['embarked'].mode().iloc[0],inplace=True)
```

```python
data['deck'].fillna(data['deck'].mode().iloc[0],inplace=True)
```

```python
data['embark_town'].fillna(data['embark_town'].mode().iloc[0],inplace=True)
```

```python
data.isnull().any()
```

# Step: 6.Find the outliers and replace the outliers.

```python
#outliers detection
sns.boxplot(data.age)
```

```python
perc99 = data.age.quantile(0.99)
perc99
```

```python
outliers = data[data.age>perc99]
outliers
```

```python
for outlier in outliers.index:
  data.loc[outlier, 'age'] = data.age.mean()
data.tail(40)
```
